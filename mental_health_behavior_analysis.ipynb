# 6. Bivariate Analysis

## (1) Daily Screen Time and Sleep Quality

```python
X = df[["Daily_Screen_Time(hrs)"]]
y = df["Sleep_Quality(1-10)"]

model = LinearRegression()
model.fit(X, y)
r2 = model.score(X, y)

sns.regplot(
    x="Daily_Screen_Time(hrs)",
    y="Sleep_Quality(1-10)",
    data=df,
    scatter_kws={"alpha": 0.6},
    line_kws={"color": "red"}
)

plt.title(f"Daily Screen Time vs Sleep Quality (R² = {r2:.2f})")
plt.xlabel("Daily Screen Time (hrs)")
plt.ylabel("Sleep Quality (1–10)")
plt.show()
```

## (2) Daily Screen Time and Stress Level

```python
X = df[["Daily_Screen_Time(hrs)"]]
y = df[["Stress_Level(1-10)"]]

model = LinearRegression()
model.fit(X, y)
r2 = model.score(X, y)

sns.regplot(
    x="Daily_Screen_Time(hrs)",
    y="Stress_Level(1-10)",
    data=df,
    scatter_kws={"alpha": 0.6},
    line_kws={"color": "red"}
)

plt.title(f"Daily Screen Time vs Stress Level (R² = {r2:.2f})")
plt.xlabel("Daily Screen Time (hrs)")
plt.ylabel("Stress Level (1–10)")
plt.show()
```

## (3) Exercise Frequency and Happiness Index

```python
sns.scatterplot(
    x="Exercise_Frequency(week)",
    y="Happiness_Index(1-10)",
    data=df
)
plt.title("Exercise Frequency vs Happiness Index")
plt.show()
```

## (4) Days Without Social Media and Happiness Index
